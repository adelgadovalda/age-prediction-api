```python
import zipfile
import os
from collections import defaultdict

# ==============================
# CONFIGURACIÓN
# ==============================
input_zip = "archive.zip"              # ZIP original
output_zip = "archive_reducido_100.zip"    # ZIP resultante
IMAGES_PER_DIR = 100                    # fotos por carpeta
VALID_EXT = (".jpg", ".jpeg", ".png")  # extensiones aceptadas
# ==============================


# Abrir zip original
with zipfile.ZipFile(input_zip, 'r') as zip_in:

    # Crear nuevo ZIP reducido
    with zipfile.ZipFile(output_zip, 'w', compression=zipfile.ZIP_DEFLATED) as zip_out:

        # Diccionario para contar imágenes por directorio
        counter = defaultdict(int)

        # Recorrer todos los elementos dentro del zip
        for info in zip_in.infolist():

            filename = info.filename

            # Saltar carpetas
            if filename.endswith('/'):
                continue

            # Verificar extensión de imagen
            if not filename.lower().endswith(VALID_EXT):
                continue

            # Obtener el directorio interno
            directory = os.path.dirname(filename)

            # Si ya hay 10 imágenes, saltamos
            if counter[directory] >= IMAGES_PER_DIR:
                continue

            # Incrementar contador
            counter[directory] += 1

            # Copiar archivo al nuevo zip
            data = zip_in.read(filename)
            zip_out.writestr(filename, data)

        print("Finalizado.")
        print(f"ZIP generado: {output_zip}")
```
